# Bubble Sort
also called shell or sinking sort algo \
*internal* , *stable* Algorithm \
subtract and conquer methodology ===>  n --> n-1, 1

---->  compare two neighboring elements, if left element > right element then swap them \
after first interation largest element will be at the rightmost index (at its correct position) 

Time complexity = Big O(n)   {best case} \
__________________Big O(n^2) {worst case}\
Number of comparisions = n-1 in first pass, n-2 in second pass -----> n(n-1)//2 total\
Number of swaps = 0 {best case}, n(n-1)/2 {worst case}

```python
arr = [6,5,4,3,2,1]

n = len(arr)
swaps = 0
for j in range(n-1): # j = 0,1,2,3,4,5  this for loop is for the number of iterations to run the algo
    flag = False   
    for i in range(n-1-j):  # n-j-1 = 0 to 5, 1 to 5 , 2 to 5 .......   this for loop is the main algo
        if arr[i] > arr[i+1]:
            arr[i], arr[i+1] = arr[i+1], arr[i]
            flag = True
            swaps += 1
    if flag == False:   # if array is already sorted than do not perform the algorithm
        break
            
print(arr)
```

# Selection Sort
 *internal*, *unstable* algorithm 
subtract and conquer ===>  n --> 1, n-1

----> take the 1st element and store it in a variable MIN, now traverse the array if current element is smaller than the MIN then store the current element in MIN, at the end of the iteration exchange the element sotred in MIN with first element then again traverse the array from 2nd element ...\
after first interation smagest element will be at the leftmost index (at its correct position) 

 
Time complexity = Big O(n^2) \
Number of comparisions = n(n-1)//2  \
Number of swaps = n-1 (in all cases)

```python
arr = [6,5,4,3,2,1]
n =len(arr)

min_index = 0 
min_value = arr[0]

for j in range(n-1):  # j = 0 1 2 3 4 
    min_index = j
    min_value = arr[j]
    for i in range(j+1,n):
        if arr[i] < min_value:
            min_value = arr[i]
            min_index = i
    arr[j], arr[min_index] = arr[min_index] , arr[j]

print(arr)
```

# Merge Sort
---> Divide and conquer \
-> divide the unsorted list in to n sublist each containing only 1 element , then repatedly mearge to produce new sorted sublist until only one sublist remain which is sorted list \
Depends on --> structure \
external , stable \
Time complexity--  best = O(n log n)  = worst

```python
def merge_sort(items):
    """
    Sorts a list of items in ascending order using the merge sort algorithm.
    """
    # The base case for the recursion: a list with 0 or 1 item is already sorted.
    if len(items) > 1:
        # Find the middle point to divide the list into two halves.
        middle = len(items) // 2
        left_half = items[:middle]
        right_half = items[middle:]

        # Recursively sort both halves.
        merge_sort(left_half)
        merge_sort(right_half)

        # Merge the sorted halves back into the original list.
        i = 0  # Pointer for the left half
        j = 0  # Pointer for the right half
        k = 0  # Pointer for the merged list

        # Compare and merge elements from both halves.
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                items[k] = left_half[i]
                i += 1
            else:
                items[k] = right_half[j]
                j += 1
            k += 1

        # Check for any remaining elements in the left half and add them.
        while i < len(left_half):
            items[k] = left_half[i]
            i += 1
            k += 1

        # Check for any remaining elements in the right half and add them.
        while j < len(right_half):
            items[k] = right_half[j]
            j += 1
            k += 1
    return items

# Example usage
my_list = [38, 27, 43, 3, 9, 82, 10]
sorted_list = merge_sort(my_list)
print(f"Sorted list is: {sorted_list}")
```
