```python
def nBinaryTriangle(n: int) -> None:
    start = 1
    for i in range(n):
        if i%2 == 0:
            start= 1
        else:
            start= 0
        for j in range(i+1):
            print(start, end=' ')
            start = 1-start
        print()
nBinaryTriangle(5)
```

```python
def numberCrown(n: int) -> None:
    space = 2 * (n - 1)
    for i in range(n):
        for j in range(i+1):
            print((j+1), end=' ')
        for j in range(space):
            print(' ', end=' ')
        for j in range(i,-1,-1):
            print((j+1), end=' ')
        print()
        space -= 2
numberCrown(5)
```

```python
def alphaHill(n: int):
    for i in range(n):
        for j in range(n- i-1):
            print(' ', end=' ')  #this is printing double spaces i.e. 2*(n-i-1) because of end=' '

        breakPoint = (2*i+1)//2
        for k in range(2*i+1):  # 2*i+1  give odd no series
            if k<= breakPoint:
                print(chr(65+k), end=' ')
            else:
                print(chr(65+breakPoint - (k- breakPoint)), end=' ')
        print()
alphaHill(4)

#both the code does same thing but both approch are diffrent
print("\nAn another way")

def alphabet_pyramid(n: int) -> None:
    for i in range(n):
        spaces = " " * (2 * (n - i - 1))  # Leading spaces
        left_part = [chr(65 + j) for j in range(i + 1)]  # Increasing sequence
        right_part = left_part[-2::-1]  # Decreasing sequence (excluding last letter)
        print(spaces + " ".join(left_part + right_part))

# Example usage
alphabet_pyramid(4)

```

```python
def getStarPattern(n: int) -> None:
    for i in range(n):
        space = n-2
        if i == 0 or i == n-1:
            print("* " * n, end=' ')
        else:
            print("* " + "  "* space + "*", end=' ')
        print()
getStarPattern(5)
```

```python
def getNumberPattern(n: int) -> None:
    for i in range(2*n-1):
        for j in range(2*n-1):
            top = i
            bottom = 2*n-2-i
            right = 2*n-2-j
            left = j
            print(n-min(top, bottom, right, left), end=' ')
        print()

getNumberPattern(3)
```
