**Contains Duplicats**<br>
Given an integer array nums, return true if any value appears more than once in the array,
otherwise return false.

Example <br>
Input: nums = [1, 2, 3, 3] <br>
Output: true

```python
def hasDuplicate(nums: list[int]) -> bool:
    parsed = set(nums)
    if len(parsed) != len(nums):
        return True
    return False

nums = [1, 2, 3, 4]
print(hasDuplicate(nums))
nums = [1, 2, 3, 4, 4]
print(hasDuplicate(nums))
```

**Valid Anagram**<br>
Given two strings s and t, return true if the two strings are anagrams of each other, otherwise return false.<br>
*An anagram is a string that contains the exact same characters as another string, but the order of the characters can be different.*

Example <br>
Input: s = "racecar", t = "carrace"<br>
Output: true

```python
from collections import Counter

def isAnagram(s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        no_of_letter_of_s = Counter(s)
        no_of_letter_of_t = Counter(t)
        if no_of_letter_of_s == no_of_letter_of_t:
            return True
        return False

s = "anagram"
t = "nagaram"
print(isAnagram(s, t))
s = "rat"
t = "car"
print(isAnagram(s, t))
```

**Two Sum**<br>
Given an array of integers nums and an integer target, return the indices i and j such that nums[i] + nums[j] == target and i != j.

You may assume that every input has exactly one pair of indices i and j that satisfy the condition.

Return the answer with the smaller index first.

Example <br>

Input: 
nums = [3,4,5,6], target = 7<br>

Output: [0,1]

```python
def twoSum(nums: list[int], target: int) -> list[int]:
        prevMap = {}

        for i , n in enumerate(nums):
            diff = target - n
            if diff in prevMap:
                return [prevMap[diff],i]
            prevMap[n] = i
        return []

nums = [2,7,11,15]
target = 9
print(twoSum(nums, target))
```
