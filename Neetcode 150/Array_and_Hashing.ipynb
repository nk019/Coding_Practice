**Contains Duplicats**<br>
Given an integer array nums, return true if any value appears more than once in the array,
otherwise return false.

Example <br>
Input: nums = [1, 2, 3, 3] <br>
Output: true

*approch* - we only need to compair the number of unique number with total numbers. so we can use set data type. 

```python
def hasDuplicate(nums: list[int]) -> bool:
    parsed = set(nums)                    # set always have any element only ones i.e. duplicate element in set is not possible
    if len(parsed) != len(nums):           # so if a list contains duplicate, its set will be of different length then it self  
        return True
    return False

nums = [1, 2, 3, 4]
print(hasDuplicate(nums))
nums = [1, 2, 3, 4, 4]
print(hasDuplicate(nums))
```

**Valid Anagram**<br>
Given two strings s and t, return true if the two strings are anagrams of each other, otherwise return false.<br>
*An anagram is a string that contains the exact same characters as another string, but the order of the characters can be different.*

Example <br>
Input: s = "racecar", t = "carrace"<br>
Output: true

*approch* - we need to count not only the number of letters in each word, but their frequencies also

```python
from collections import Counter

def isAnagram(s: str, t: str) -> bool:
    no_of_letter_of_s = Counter(s)            # counter returns a dict with key (each unique letter)  : value (frequency of that letter)      
    no_of_letter_of_t = Counter(t)            
    if no_of_letter_of_s == no_of_letter_of_t:
        return True
    return False

s = "anagram"
t = "nagaram"
print(isAnagram(s, t))
s = "rat"
t = "car"
print(isAnagram(s, t))
```

**Two Sum**<br>
Given an array of integers nums and an integer target, return the indices i and j such that nums[i] + nums[j] == target and i != j.<br>
You may assume that every input has exactly one pair of indices i and j that satisfy the condition.<br>
Return the answer with the smaller index first.

Example <br>
Input: 
nums = [3,4,5,6], target = 7<br>
Output: [0,1]

*approch* - for each element we search for it compliment to be present i.e. TARGET - element

```python
def twoSum(nums: list[int], target: int) -> list[int]:
    prevMap = {}                                      # we will store list element here with their index as their value

    for index , num in enumerate(nums):          # i will be index, n will be the element    key:value ==>  element:index
        diff = target - num               # diff is the compliment for which we will search in our dict/ hashmap
        if diff in prevMap: 
            return [prevMap[diff],index]         #if diff is found in hashmap then we will fetch indices for diff and that element (i)
        prevMap[num] = index                    # if diff is not found then we will store that element in hashmap.
    return []               

nums = [12,7,11,1]
target = 12
print(twoSum(nums, target))
```

**Group Anagrams**<br>
Given an array of strings strs, group all anagrams together into sublists. You may return the output in any order.<br>
An anagram is a string that contains the exact same characters as another string, but the order of the characters can be different.


```python
from collections import defaultdict

def groupAnagrams(strs: list[str]) -> list[list[str]]:
        
    res = defaultdict(list)               # normal dict me, agar koi key na ho or use call karo to key error aa jata h but default dict me nhi aata 
    for s in strs:
        count = [0] * 26                  # a to z  har str me kitne letter hain unko count karne ke liye 
        for c in s:
            count[ord(c)-ord("a")] += 1   #  us count ko key banaya   ord retruns unicode code(integer) representing a character
            
        res[tuple(count)].append(s)       # key ko unmutable banane ke liye tuple and uski value str ko res (dict) me append kar diya
        
    return list(res.values())
    
strs = ["act","pots","tops","cat","stop","hat"]
print(groupAnagrams(strs))
```

**Top k frequent elements**<br>
Given an integer array nums and an integer k, return the k most frequent elements within the array.<br>
The test cases are generated such that the answer is always unique.<br>
You may return the output in any order.<br>

```python
"""
we take different approch for the same problem 
we create dict for unique element and their freq using counter. then sort that dict by values (frequencies are stored as values) in reverse order 
iterate over the values and add them in a list , then retrun first 'k' elements of that list 
"""
def topKFrequent(nums: list[int], k: int) -> list[int]:
        hashmap = dict(Counter(nums))                 # this creats the frequency map (dictionary)
        sorted_hash = dict(sorted(hashmap.items(), key = lambda item: item[1], reverse=True))   # sort by value (frequency) in decending order
        result = []                           
        for i in sorted_hash.keys():          # we iterate over key because keys are the actual element we want
            result.append(i)
        return result[0:k]               # top k elemnt in list are our result

nums = [1,2,2,3,3,3]; k = 2
print(topKFrequent(nums,k))
```

**Encode and Decode Strings** <br>
Design an algorithm to encode a list of strings to a string. The encoded string is then sent over the network and is decoded back to the original list of strings.<br>


```python
def encode(strs: list[str]) -> str:    # encodeing is simple - we create an empty string and add each string of list in that 
        result = ""                    # empty string with this format --> length(string)#string
        for s in strs:
            result += str(len(s))+'#'+ s
        return result

def decode(s: str) -> list[str]:
    decode = []                      # we use two pointer, whenever we encounter _#_ symbol we copy characters after 
    i = 0
    while i < len(s):
        j = i
        while s[j] != '#':
            j += 1
        length = int(s[i:j])           #   why i:j   why not just i??????????  ---- some test cases were not passing
        i= j + 1
        j = i + length
        decode.append(s[i:j])
        i = j

    return decode

dummy_input=["Hello","World"]
print(encode(dummy_input))
print(decode(encode(dummy_input)))
```

**Products of array except itself**<br>
Given an integer array nums, return an array output where output[i] is the product of all the elements of nums except nums[i].<br>
Each product is guaranteed to fit in a 32-bit integer.<br>
Follow-up: Could you solve it in O(n) time without using the division operation?

*Approch* -<br>
- prefix is product of all previous element in for an element i. (while going form left to right)<br>
- postfix is proudct of all element that will come after i. (or previous elements if we go from right to left)<br>
- First we go from left to right and for each element we store its prefix in output array. then we go from right to left<br>
  and for each element we multipy (*not addition*) the postfix to output array.
   

```python
def productExceptSelf(nums: list[int]) -> list[int]:
    N= len(nums)
    output = [1] * N
    prefix = 1
    for i in range(N):
        output[i] = prefix
        prefix *= nums[i]
    postfix = 1
    for i  in range(N-1, -1 , -1):
        output[i] *= postfix
        postfix *= nums[i]
    return output 
nums = [1,2,4,6]
productExceptSelf(nums)
```

**Valid sudoku**<br>
You are given a 9 x 9 Sudoku board board. A Sudoku board is valid if the following rules are followed:<br>
- Each row must contain the digits 1-9 without duplicates.
- Each column must contain the digits 1-9 without duplicates.
- Each of the nine 3 x 3 sub-boxes of the grid must contain the digits 1-9 without duplicates.<br>
Return true if the Sudoku board is valid, otherwise return false

Note: A board does not need to be full or be solvable to be valid.

```python
def isValidSudoku(self, board: List[List[str]]) -> bool:
    rows = [0] * 9
    cols = [0] * 9
    squares = [0] * 9

    for r in range(9):
        for c in range(9):
            if board[r][c] == ".":
                continue

            val = int(board[r][c]) - 1         # maping all 9 digit to bits 0 - 8
            if (1 << val) & rows[r]:
                return False
            if (1 << val) & cols[c]:
                return False
            if (1 << val) & squares[(r // 3) *3 + (c // 3)]:
                return False
                
            rows[r] |= (1 << val)
            cols[c] |= (1 << val)
            squares[(r // 3) * 3 + (c // 3)] |= (1 << val)
                
    return True   
```
